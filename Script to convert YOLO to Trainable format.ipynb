# Script to generate training data for YOLO.

```python
import os
import cv2
from xml.dom.minidom import parseString
from lxml.etree import Element, SubElement, tostring
import numpy as np
from os.path import join

YOLO_CLASSES = ('Cup')
#classid for cup is 0 and box is 1
class_id = 0

folder_path = "C:/Users/Yogesh/CupDetector/images/"

# converts the normalized positions  into integer positions
def unconvert(class_id, width, height, x, y, w, h):

    xmax = int((x*width) + (w * width)/2.0)
    xmin = int((x*width) - (w * width)/2.0)
    ymax = int((y*height) + (h * height)/2.0)
    ymin = int((y*height) - (h * height)/2.0)
    class_id = int(class_id)
    return (class_id, xmin, ymin, xmax, ymax)
    
# Main
#write to a text file
write_file = open("cup_train.txt", "w+")

list_dir = os.listdir(folder_path)
number_of_files = 0
for file in list_dir:
    if file.endswith('.jpg'):
        number_of_files += 1    

for i in range (0, number_of_files):
    yolo_format_txt_path = folder_path+"cup"+str(i)+".txt"
    img_path = folder_path+"cup"+str(i)+".jpg"
    string = " "
    
    if os.path.exists(yolo_format_txt_path):
        file = open(yolo_format_txt_path, "r")
        string = img_path
        contents = file.read().split("\n")
        for i, content in enumerate(contents):
            if len(content) > 0: 
                temp = content.split(" ")
                label = unconvert(0, 10, 10, float(temp[1]), float(temp[2]), float(temp[3]), float(temp[4]))
                string+= " "+str(label[1])+","+str(label[2])+","+str(label[3])+","+str(label[4])+","+str(label[0])

    write_file.write(string+"\n")
    print(string)
```
